```python
df['INFP'] = df.posts.str.count('INFP')
df['INFJ'] = df.posts.str.count('INFJ')
df['INTP'] = df.posts.str.count('INTP')
df['INTJ'] = df.posts.str.count('INTJ')
df['ENTP'] = df.posts.str.count('ENTP')
df['ENFP'] = df.posts.str.count('ENFP')
df['ISTP'] = df.posts.str.count('ISTP')
df['ISFP'] = df.posts.str.count('ISFP')
df['ENTJ'] = df.posts.str.count('ENTJ')
df['ISTJ'] = df.posts.str.count('ISTJ')
df['ENFJ'] = df.posts.str.count('ENFJ')
df['ISFJ'] = df.posts.str.count('ISFJ')
df['ESTP'] = df.posts.str.count('ESTP')
df['ESFP'] = df.posts.str.count('ESFP')
df['ESFJ'] = df.posts.str.count('ESFJ')
df['ESTJ'] = df.posts.str.count('ESTJ')
```

```python
df
```

```python
df_corr = df.groupby(['type'])[["INFP", "INFJ", "INTP", "INTJ", 
                      "ENTP", "ENFP", "ISTP", "ISFP", 
                      "ENTJ", "ISTJ", "ENFJ", "ISFJ", 
                      "ESTP", "ESFP", "ESFJ", "ESTJ"]].sum()
df_corr.index.name = None
df_corr = df_corr.transpose().reset_index()
df_corr.rename(columns = {'index': 'type'}, inplace=True)
```

```python
df_corr
```

```python
type_list = df_corr.columns.get_values()[1:]
```

```python
sns.set(font_scale=1.25)
sns.set_style("whitegrid")
fig, ax = plt.subplots(nrows=4, ncols=4, figsize=(16,12))
fig.tight_layout(h_pad=4)
num_plots = 16

for i in range(num_plots):
    act_type = type_list[i]
    type_df = df_corr[['type', act_type]]
    row = i // 4
    col = i % 4
    ax_curr = ax[row, col]
    clrs = ['grey' if (x < max(type_df[act_type])) else 'red' for x in type_df[act_type] ]
    act_barplot = sns.barplot(x="type", y=act_type, data=type_df, palette=clrs, ax=ax_curr)
    if col == 0:
        ax_curr.set_ylabel('Occurance')
    else:
        ax_curr.set_ylabel('')
    ax_curr.set_title(act_type)
    ax_curr.set_xlabel('')
    ax_curr.tick_params(axis='x', which='major', pad=2)
    for item in act_barplot.get_xticklabels():
        item.set_rotation(90)
    ax_curr.get_yaxis().set_label_coords(-0.22,0.5)
        
plt.savefig('subplots.png', format='png', dpi=100, bbox_inches='tight')
```
